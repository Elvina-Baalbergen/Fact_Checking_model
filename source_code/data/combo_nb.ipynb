```python
data = pd.read_csv('../../data/raw/MovieSummaries/movie.metadata.tsv', delimiter='\t', header=None, names = ['drop1','index','title', 'year', 'drop2', 'drop3', 'drop4','drop5','genre'])
print(len(data))
```

```python
metadata = data.drop(['drop1','drop2','drop3','drop4','drop5'], axis=1)
print(metadata.head(16))
```

```python
# Check which rows have at least one missing value in any column
missing_data = metadata.isnull().any(axis=1)

# Print the rows with missing data
print(metadata[missing_data])
```

```python
metadata.dropna(inplace=True)
print(len(metadata))
```

```python
metadata['year'] = metadata['year'].astype(str).str[:4]

print(metadata.head(7))
```
